# Whole-Brain Parcellation (Schaefer+Tian) Results Viewer

This notebook loads result `.joblib` files and `.png` figures from:

`/Users/xiaoqianxiao/projects/NARSAD/LSS/results/wholebrain_parcellation_schaefer`


```
from pathlib import Path
import joblib
import pandas as pd
import numpy as np

RESULTS_DIR = Path("/Users/xiaoqianxiao/projects/NARSAD/LSS/results/wholebrain_parcellation_schaefer")
print("Results dir:", RESULTS_DIR)
print("Exists:", RESULTS_DIR.exists())
```

```
# List available result files
joblib_files = sorted(RESULTS_DIR.glob("*.joblib"))
fig_files = sorted(RESULTS_DIR.glob("*.png"))

print("Joblib files:", len(joblib_files))
for f in joblib_files:
    print(" -", f.name)

print("
Figure files:", len(fig_files))
for f in fig_files:
    print(" -", f.name)
```

```
# Quick loader utility

def load_joblib(name):
    path = RESULTS_DIR / name
    if not path.exists():
        raise FileNotFoundError(path)
    return joblib.load(path)
```

## Analysis 11 (Neural Dissociation)


```
res11 = load_joblib("results_11.joblib")
res11.keys()
```

```
# Print key stats
print("acc_sad_cv:", res11.get("acc_sad_cv"))
print("acc_hc_cv:", res11.get("acc_hc_cv"))
print("p_sad:", res11.get("p_sad"))
print("p_hc:", res11.get("p_hc"))
print("p_sim:", res11.get("p_sim"))
```

```
# Display figures if present
from IPython.display import Image, display

for fname in ["analysis_11_neural_dissociation.png", "results_11_neural_dissociation.png"]:
    f = RESULTS_DIR / fname
    if f.exists():
        display(Image(filename=str(f)))
    else:
        print("Missing:", f.name)
```

## Analysis 12–27 Outputs


```
# Load and summarize selected results
summary = {}
for f in joblib_files:
    name = f.name
    try:
        obj = joblib.load(f)
        if isinstance(obj, dict):
            summary[name] = list(obj.keys())[:10]
        else:
            summary[name] = type(obj).__name__
    except Exception as e:
        summary[name] = f"ERROR: {e}"

pd.DataFrame.from_dict(summary, orient="index", columns=["preview"]).sort_index()
```
